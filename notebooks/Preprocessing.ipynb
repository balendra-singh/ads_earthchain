```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('../data/staging/data.csv')

data['size'].value_counts()
```

```python
cols = data.filter(like='Goal_').columns
data[cols] = cols.fillna(False)
data.isna().sum()
```

```python
data.replace(['Microscale','Large scale','Micro scale'],['Micro Scale','Large Scale','Micro Scale'],inplace=True)
data['num_days']=(pd.to_datetime(data['crediting_period_end_date'])-pd.to_datetime(data['crediting_period_start_date'])).dt.days
data.loc[data['VER_retired_credits']>data['VER_issued_credits'],'VER_retired_credits']=data['VER_issued_credits']
data['VER_percentage_sold']=data['VER_retired_credits']/data['VER_issued_credits']
data['VER_percentage_sold_per_day']=data['VER_percentage_sold']/data['num_days']
```

```python
import pycountry_convert as pc
data['continent_code']=data['country_code'].replace(['TL','XZ'],['TP','US']).apply(pc.country_alpha2_to_continent_code)
data.loc[data['country_code']=='XZ','continent_code']='XZ'
```

```python
data.iloc[:,25:41].fillna(False,inplace=True)
data['num_of_goals']=np.sum(data.iloc[:,25:41],axis=1)
```

```python
import sklearn.preprocessing as preprocessing
```

```python
CountryEncoder=preprocessing.OneHotEncoder(sparse=False)
CountryEncoder.fit(data[['country_code']])
data[CountryEncoder.get_feature_names_out()]=CountryEncoder.transform(data[['country_code']])
ContinentEncoder=preprocessing.OneHotEncoder(sparse=False)
ContinentEncoder.fit(data[['continent_code']])
data[ContinentEncoder.get_feature_names_out()]=ContinentEncoder.transform(data[['continent_code']])
SizeEncoder=preprocessing.OneHotEncoder(sparse=False)
SizeEncoder.fit(data[['size']])
data[SizeEncoder.get_feature_names_out()]=SizeEncoder.transform(data[['size']])
data
```

```python
data.to_csv('prepped_data.csv')
```
