```python
import http.client
import json
import pprint



gs_certified_project_list = []

conn = http.client.HTTPSConnection("api.goldstandard.org")
payload = ''
headers = {
'Content-Type': 'application/json'
}

page_counter = 1

while(True):
  conn.request("GET", f"/projects?size=150&page={page_counter}&gsCertified=true", payload, headers)
  res = conn.getresponse()
  response_json =  json.loads(res.read().decode())

  if(len(response_json) == 0):
    print(f"Data retrieved till page number {page_counter}")
    break;

  gs_certified_project_list += response_json
  page_counter += 1
  #pd.DataFrame(list(zip(lst1, lst2, lst3)))

print(f"Total Gold Standard Certified Project Found are {len(gs_certified_project_list)}")

```
