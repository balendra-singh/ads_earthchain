### Dropping unnecessary columns from the dataframe

```python
df_corr = indicator_df.iloc[:,1:].corr().abs() 

#Converting correlation matrix in a heatmap. 
df_corr.style.background_gradient(cmap='coolwarm',axis=None).format(precision=2) 


masks = np.zeros_like(df_corr, dtype=bool)
masks[np.triu_indices_from(masks)] = True 
#To avoid redundancy of the matrix (no repeated correlation coefficients)

df_corr[masks] = np.nan 

# Because there is error on null_color hyper parameter due to different pandas version, 
# we used try-except to make it work in any version.
try:
         df_corr_styled=(df_corr
                       .style
                       .format(precision=2)
                       .background_gradient(cmap='coolwarm', axis=None, vmin=0, vmax=1)
                       .highlight_null(null_color='#f1f1f1'))
except: 
         df_corr_styled=(df_corr
                       .style
                       .format(precision=2)
                       .background_gradient(cmap='coolwarm', axis=None, vmin=0, vmax=1)
                       .highlight_null(color='#f1f1f1'))
```

### Transforming the categorically encoded data to evaluate score

```python
scaler = StandardScaler()
cleaned_df.iloc[:,48:] = scaler.fit_transform(cleaned_df.iloc[:,48:].to_numpy())
cleaned_df.iloc[:,48:] = np.abs(cleaned_df.iloc[:,48:])
```

```python
scaler_d = StandardScaler()
cleaned_df[['crediting_days', 'VER_sold_percentage']] = np.abs(scaler.fit_transform(cleaned_df[['crediting_days', 'VER_sold_percentage']]))
```

### Creating A Score

Goal 7, Goal 12 and Goal 13 impact the most to the carbon emissions reduction based on the description of the SDGs. Hence, the weights assigned to those are high. 
Additionally, from the heatmap, it can be presumed that Methodology and Size impact the VER_sold_percentage. Hence, those have been assigned higher weightage. 

The rest of the chosen features just impact the VER_sold_percentage and not necessarily have very high correlation values. 

```python
cleaned_df['score'] = 2 * (cleaned_df['Goal_7'] + cleaned_df['Goal_12'] + cleaned_df['Goal_13']) + cleaned_df['gsf_std_version_enc'] + \
    2 * (cleaned_df['methodology_enc'] + cleaned_df['size_enc']) + cleaned_df['type_enc'] + cleaned_df['proj_dev_enc'] + cleaned_df['country_enc'] + cleaned_df['crediting_days'] + \
        cleaned_df['Goal_1'] + cleaned_df['Goal_2'] + cleaned_df['Goal_3'] + cleaned_df['Goal_4'] + cleaned_df['Goal_5'] + cleaned_df['Goal_6'] + cleaned_df['Goal_8'] + \
            cleaned_df['Goal_9'] + cleaned_df['Goal_10'] + cleaned_df['Goal_11'] + cleaned_df['VER_sold_percentage']
```
